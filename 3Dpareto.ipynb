```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

```python
pareto_Data = pd.ExcelFile('3DParetoData.xlsx')
```

```python
df_P1S2 = pd.read_excel(pareto_Data, 'P1S2')
df_P2S2 = pd.read_excel(pareto_Data, 'P2S2')
df_P3S2 = pd.read_excel(pareto_Data, 'P3S2')
df_P4S2 = pd.read_excel(pareto_Data, 'P4S2')
df_P5S2 = pd.read_excel(pareto_Data, 'P5S2')
```

```python
x_P1S2 = (df_P1S2['p_cost']).to_numpy().T
y_P1S2 = (df_P1S2['p_time']).to_numpy().T
z_P1S2 = (df_P1S2['makespan']).to_numpy().T

```

```python
# Creating figure
fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(x_P1S2, y_P1S2, z_P1S2, color = "green")
plt.title("P1S2")
ax.set_xlabel('p_cost', fontweight ='bold')
ax.set_ylabel('p_time', fontweight ='bold')
ax.set_zlabel('makespan', fontweight ='bold')
plt.savefig("P1S2.jpg")

```
